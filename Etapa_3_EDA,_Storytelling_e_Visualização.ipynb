```python
# Importar Pandas
import pandas as pd

# Importar RE - Express√µes Regulares
import re
```

```python
dataframe_final = pd.read_csv("dataframe_final_candidatos.csv")
```

```python
dataframe_final.head()
```

```python
# Identificar tipos dos dados
dataframe_final.info()
```

```python
# Somar todos os 'nulos', quando houver.
dataframe_final.isnull().sum()
```

```python
# Criar um coluna 'Tweet_limpo' que ser√° a coluna final ap√≥s a limpeza dos dados.
dataframe_final['Tweet_limpo'] = dataframe_final['Tweet']
```

```python
dataframe_final.head()
```

```python
# Criar um fun√ß√£o que ir√° realizar alguns tratamento na coluna 'Tweet' e salvar na coluna 'Tweet_Limpo'
def limpeza_do_texto(tweet):
  # Deixa todos os tweet min√∫sculos.
  tweet = tweet.lower()

  tweet = re.sub('\n', '', tweet)
  tweet = re.sub('@[^\s]+', '', tweet)
  tweet = re.sub('\B#\S+', '', tweet)
  tweet = re.sub('https?://\S+|www\.\S+', '', tweet)
  return tweet
```

```python
# Referencia a coluna origem 'Tweet' onde ser√° aplicado a fun√ß√£o 'limpeza_do_texto' e o resultado ser√° gravado na coluna 'Tweet_Limpo'.
dataframe_final['Tweet_limpo'] = dataframe_final['Tweet'].apply(lambda i: limpeza_do_texto(i))
```
